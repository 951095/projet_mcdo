```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger les données
df = pd.read_csv('ObesityDataSet_raw_and_data_sinthetic.csv')

# Sélectionner uniquement les caractéristiques nécessaires
X = df[["Gender", "Age", "Height", "SMOKE", "family_history_with_overweight", "FAVC"]]
y = df["NObeyesdad"]

# Graphique 1: Répartition des classes cibles (NObeyesdad)
plt.figure(figsize=(8, 6))
sns.countplot(x=y, data=df)
plt.title('Répartition des classes cibles (NObeyesdad)')
plt.xlabel('Classes')
plt.ylabel("Nombre d'individus")
plt.xticks(rotation=45)
plt.show()

# Graphique 2: Répartition des genres (Gender)
plt.figure(figsize=(6, 4))
sns.countplot(x=X["Gender"], data=df)
plt.title('Répartition des genres (Gender)')
plt.xlabel('Genre')
plt.ylabel("Nombre d'individus")
plt.xticks(ticks=[0, 1], labels=['Female', 'Male'])
plt.show()

# Graphique 3: Histogramme de l'âge (Age)
plt.figure(figsize=(8, 6))
sns.histplot(X['Age'], bins=20, kde=True)
plt.title('Histogramme de l\'âge (Age)')
plt.xlabel('Âge')
plt.ylabel("Nombre d'individus")
plt.show()

# Graphique 4: Répartition de la taille (Height)
plt.figure(figsize=(8, 6))
sns.histplot(X['Height'], bins=20, kde=True)
plt.title('Répartition de la taille (Height)')
plt.xlabel('Taille')
plt.ylabel("Nombre d'individus")
plt.show()

# Graphique 5: Répartition de la consommation de tabac (SMOKE)
plt.figure(figsize=(6, 4))
sns.countplot(x=X["SMOKE"], data=df)
plt.title('Répartition de la consommation de tabac (SMOKE)')
plt.xlabel('Tabagisme')
plt.ylabel("Nombre d'individus")
plt.xticks(ticks=[0, 1], labels=['Non fumeur', 'Fumeur'])
plt.show()

# Graphique 6: Répartition des antécédents familiaux d'excès de poids (family_history_with_overweight)
plt.figure(figsize=(6, 4))
sns.countplot(x=X["family_history_with_overweight"], data=df)
plt.title('Répartition des antécédents familiaux d\'excès de poids')
plt.xlabel('Antécédents familiaux')
plt.ylabel("Nombre d'individus")
plt.xticks(ticks=[0, 1], labels=['Non', 'Oui'])
plt.show()

# Graphique 7: Répartition de la consommation de fast food (FAVC)
plt.figure(figsize=(6, 4))
sns.countplot(x=X["FAVC"], data=df)
plt.title('Répartition de la consommation de fast food (FAVC)')
plt.xlabel('Consommation de fast food')
plt.ylabel("Nombre d'individus")
plt.xticks(ticks=[0, 1], labels=['Non', 'Oui'])
plt.show()

# Graphique 1: Diagramme en barres pour Gender
plt.figure(figsize=(8, 6))
sns.countplot(x="Gender", hue="NObeyesdad", data=df)
plt.title('Répartition de NObeyesdad en fonction du Genre')
plt.xlabel('Genre')
plt.ylabel("Nombre d'individus")
plt.xticks(ticks=[0, 1], labels=['Female', 'Male'])
plt.legend(title='NObeyesdad', loc='upper right')
plt.show()

# Graphique 2: Diagramme en barres pour SMOKE
plt.figure(figsize=(8, 6))
sns.countplot(x="SMOKE", hue="NObeyesdad", data=df)
plt.title('Répartition de NObeyesdad en fonction du Tabagisme')
plt.xlabel('Tabagisme')
plt.ylabel("Nombre d'individus")
plt.xticks(ticks=[0, 1], labels=['Non fumeur', 'Fumeur'])
plt.legend(title='NObeyesdad', loc='upper right')
plt.show()

# Graphique 3: Boîte à moustaches pour Age et Height
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x="NObeyesdad", y="Age", data=df)
plt.title('Distribution de l\'âge par classe de NObeyesdad')
plt.xlabel('NObeyesdad')
plt.ylabel('Âge')

plt.subplot(1, 2, 2)
sns.boxplot(x="NObeyesdad", y="Height", data=df)
plt.title('Distribution de la taille par classe de NObeyesdad')
plt.xlabel('NObeyesdad')
plt.ylabel('Taille')

plt.tight_layout()
plt.show()

# Graphique 4: Diagramme en barres pour family_history_with_overweight et FAVC
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.countplot(x="family_history_with_overweight", hue="NObeyesdad", data=df)
plt.title('Répartition de NObeyesdad en fonction des antécédents familiaux d\'excès de poids')
plt.xlabel('Antécédents familiaux d\'excès de poids')
plt.ylabel("Nombre d'individus")
plt.xticks(ticks=[0, 1], labels=['Non', 'Oui'])
plt.legend(title='NObeyesdad', loc='upper right')

plt.subplot(1, 2, 2)
sns.countplot(x="FAVC", hue="NObeyesdad", data=df)
plt.title('Répartition de NObeyesdad en fonction de la consommation de fast-food')
plt.xlabel('Consommation de fast-food')
plt.ylabel("Nombre d'individus")
plt.xticks(ticks=[0, 1], labels=['Non', 'Oui'])
plt.legend(title='NObeyesdad', loc='upper right')

plt.tight_layout()
plt.show()

# Graphique 5: Nuage de points pour Age et Height
plt.figure(figsize=(8, 6))
sns.scatterplot(x="Age", y="Height", hue="NObeyesdad", data=df)
plt.title('Nuage de points de l\'âge et de la taille, colorés selon la classe de NObeyesdad')
plt.xlabel('Âge')
plt.ylabel('Taille')
plt.legend(title='NObeyesdad', loc='upper right')
plt.show()
```
